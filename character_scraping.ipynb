# WEB SCRAPING - GAME OF THRONES characters

### OBJECTIVE:
Extracting an online list of 'game of thrones' characters into a csv file

```python
# IMPORTING LIBRARIES

#Selenium is used for web-automation purposes
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

import time
import pandas as pd
import numpy as np
```

```python
driver = webdriver.Chrome()
characters_url = "https://en.wikipedia.org/wiki/List_of_Game_of_Thrones_characters"
driver.get(characters_url)


# time.sleep(2)
# driver.find_element(By.XPATH, "//div[text()='ACCETTA TUTTI']").click()
```

```python
# Iterate over the wikipedia table and add elements contained in cells in the t_characters list
got_characters_elem = driver.find_elements(By.XPATH, '//table[@class = "wikitable col1left col2left sortable jquery-tablesorter"]/tbody/tr/td[2]')
got_characters = [character.text for character in got_characters_elem]
```

```python
# Some characters in the wikipedia table are found as "firstName 'Alias' surname"
# so i split the complete name adding aliasses to a different column
aliases = []
character_names = []
for character in got_characters:
    if character.find('"') != -1:
        aliases.append(character[character.find('"')+1:character.rfind('"')])
        character.replace('"', '')
        character_names.append(character[:character.find('"')] + character[character.rfind('"')+1:])
    else:
        character_names.append(character)
        aliases.append(np.nan)


data = {'Name': character_names, 'Alias': aliases}
df = pd.DataFrame(data = data)
df
```

```python
# Save the dataframe to a csv file
df.to_csv('character_list.csv', header='true', index = False)
```
